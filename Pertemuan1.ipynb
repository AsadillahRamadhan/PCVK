```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as pit
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print('resolusi image: tinggi x lebar = ', tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
    image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#Garis Vertikal dan Diagonal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi):
    image_3[x,int((lebar)/2)] = [255,255,255]

gap = lebar - tinggi
add = tinggi % gap
divider =  round(tinggi / gap)
addon = tinggi % gap
gap_1 = round(tinggi / divider)
gap_2 = round(tinggi / (tinggi / gap))
gap_3 = gap_2 - gap_1
sisa = round(tinggi / gap_3)
px = 0

for x in range (tinggi):
    image_4[x,px] = [255,255,255]
    if(x % divider == 0):
      if(x % sisa == 0):
        if(px >= lebar - 1):
         break
        image_4[x,px + 1] = [255,255,255]
        px += 1
      if(px >= lebar - 1):
        break
      image_4[x,px + 1] = [255,255,255]
      px += 1
    px += 1

px = 0
for x in range (tinggi - 1, -1, -1):
    image_4[x,px] = [255,255,255]
    if(x % divider == 0):
      if(x % sisa == 0):
        if(px >= lebar - 1):
         break
        image_4[x,px + 1] = [255,255,255]
        px += 1
      if(px >= lebar - 1):
        break
      image_4[x,px + 1] = [255,255,255]
      px += 1
    px += 1

final_frame_1 = cv.hconcat((image_2, image_3))
final_frame_2 = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame_1)
print('\n')
cv2_imshow(final_frame_2)
```

```python
#Garis Horizontal Dibagian Tengah Gambar Dengan Panjang Tertentu
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(150, 200, 1):
  image_5[int(tinggi / 2.5),y] = [255,255,255]



final_frame_2 = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame_2)
```

```python
#Membuat Kotak Dalam Gambar
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(150, 200, 1):
  for x in range(50):
    image_6[y,x+ int(lebar / 3)] = [255,255,255]



final_frame_2 = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame_2)
```
