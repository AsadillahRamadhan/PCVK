```python
# Transformasi Linier Brightness

import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print(' Mengubah tingkat kecerahan citra ')
print('----------------------------------')
try:
    brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/female.jpeg')
brightness_image = np.zeros(original.shape, original.dtype)

# akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```python
# Tugas Praktikum 1
baboon_1 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/baboon.png')
negative = np.zeros(baboon_1.shape, baboon_1.dtype)

for y in range(baboon_1.shape[0]):
  for x in range(baboon_1.shape[1]):
    for c in range(baboon_1.shape[2]):
      negative[y,x,c] = np.clip(255 - baboon_1[y,x,c], 0, 255)

final_frame = cv.hconcat((baboon_1, negative))
cv2_imshow(final_frame)
```

```python
# Transformasi Contrast

print(' Mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------------------')
try:
    brightness = int(input('Masukkan nilai kecerahan [-255 - 255]: '))
    contrast = int(input('Masukkan kontras[1.0 - 3.0]: '))
except ValueError:
    print('Error, not a number')

female_2 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/female.jpeg')
brightness_and_contrast = np.zeros(female_2.shape, female_2.dtype)

for y in range(female_2.shape[0]):
  for x in range(female_2.shape[1]):
    for c in range(female_2.shape[2]):
      brightness_and_contrast[y,x,c] = np.clip(contrast * female_2[y,x,c] + brightness, 0, 255)

final_frame = cv.hconcat((female_2, brightness_and_contrast))
cv2_imshow(final_frame)
```

```python
# Logarithmic Brightness

print(' Mengubah tingkat kecerahan citra dengan Transformasi Log ')
print('----------------------------------------------------------')
try:
    brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, not a number')

female_3 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/female.jpeg')
logarithmic_brightness = np.zeros(female_3.shape, female_3.dtype)

for y in range(female_3.shape[0]):
      logarithmic_brightness[y] = np.clip(brightness * np.log(1 + female_3[y]), 0, 255)

final_frame = cv.hconcat((female_3, logarithmic_brightness))
cv2_imshow(final_frame)
```

```python
# Averaging

baboon_3 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/baboon.png')
averaging = np.zeros(baboon_3.shape, baboon_3.dtype)

for y in range(baboon_3.shape[0]):
  for x in range(baboon_3.shape[1]):
    tes = 0
    color = 0
    scale = [0, 0, 0]
    for c in range(baboon_3.shape[2]):
      scale[c] = baboon_3[y,x,c]
      tes += scale[c]
      for a in range(3):
        color = tes / 3
        averaging[y, x, a] = np.clip(color, 0, 255)


final_frame = cv.hconcat((baboon_3, averaging))
cv2_imshow(final_frame)
```

```python
# Lightness

baboon_4 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/baboon.png')
lightness = np.zeros(baboon_4.shape, baboon_4.dtype)

for y in range(baboon_4.shape[0]):
  for x in range(baboon_4.shape[1]):
    max = 0
    min = 255
    color = 0
    scale = [0, 0, 0]
    for c in range(baboon_4.shape[2]):
      if(baboon_4[y,x,c] > max):
        max = baboon_4[y,x,c]
      if(baboon_4[y,x,c] < min):
        max = baboon_4[y,x,c]
      for a in range(3):
        color = (max + min) / 2
        lightness[y, x, a] = np.clip(color, 0, 255)


final_frame = cv.hconcat((baboon_4, lightness))
cv2_imshow(final_frame)
```

```python
# Luminance

baboon_5 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/baboon.png')
luminance = np.zeros(baboon_5.shape, baboon_5.dtype)

for y in range(baboon_5.shape[0]):
  for x in range(baboon_5.shape[1]):
    color = 0
    divider = [0.21, 0.72, 0.07]
    for c in range(baboon_5.shape[2]):
      color += baboon_5[y,x,c] * divider[c]
      for a in range(3):
        luminance[y, x, a] = np.clip(color, 0, 255)


final_frame = cv.hconcat((baboon_5, luminance))
cv2_imshow(final_frame)
```

```python
# Menampilkan Warna Tertentu Pada Citra

focus = input('Masukkan nilai yang ingin difokuskan (R, G, B): ')

baboon_5 = cv.imread('/content/drive/MyDrive/PCVK/Pertemuan3/baboon.png')
focused_image = np.zeros(baboon_5.shape, baboon_5.dtype)

try:
  if(focus == 'R'):
    code = 2
  elif(focus == 'G'):
    code = 1
  elif(focus == 'B'):
    code = 0
except ValueError:
  print('Masukkan Nilai Yang Benar!')

for y in range(baboon_5.shape[0]):
    for x in range(baboon_5.shape[1]):
      tes = 0
      color = 0
      scale = [0, 0, 0]
      max = 0
      maxscale = 0
      for c in range(baboon_5.shape[2]):
        scale[c] = baboon_5[y,x,c]
        tes += scale[c]

        if(baboon_5[y,x,c] > max):
          maxscale = c
          max = baboon_5[y,x,c]

      if(maxscale == code):
        for a in range(baboon_5.shape[2]):
          focused_image[y, x, a] = np.clip(baboon_5[y,x,a], 0, 255)
      else:
        color = tes / 3
        for a in range(baboon_5.shape[2]):
          focused_image[y, x, a] = np.clip(color, 0, 255)

final_frame = cv.hconcat((baboon_5, focused_image))
cv2_imshow(final_frame)
```
